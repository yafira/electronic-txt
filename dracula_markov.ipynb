# markov txt generation from *dracula* by bram stoker
using shoestrings by Allison Parrish, https://github.com/aparrish/shoestrings/tree/main

```python
import re
import random
import textwrap
from shoestrings import TextGenerator
from shoestrings.tokenizers import SimpleWordTokenizer

with open('dracula.txt', 'r', encoding='utf-8') as f:
    raw = f.read()

text = re.sub(r'\s+', ' ', raw).strip()
sentences = [s.strip() for s in re.split(r'(?<=[.!?])\s+', text) if len(s.strip()) > 20]

print(f"{len(text.split())} words, {len(sentences)} sentences")
```

## markov default
order-2: the model looks at the 2 words before each word to decide what comes next. <br>
enough context to sound like stoker's diary entries, but the logic still breaks down.

```python
tok = SimpleWordTokenizer()
gen_default = TextGenerator(n=3, tokenizer=tok, source=sentences)

for _ in range(4):
    print(textwrap.fill(gen_default.generate_one(max_tokens=50, temperature=1.5), width=80))
    print()
```

## glitchy
order-1: only looks back 1 word. <br>
barely any context, so the syntax collapses. what's left is dracula's most repeated words looping without meaning ex: blood, night, count, fear.

```python
tok_glitch = SimpleWordTokenizer()
gen_glitchy = TextGenerator(n=2, tokenizer=tok_glitch, source=[text])

words = gen_glitchy.generate_one(max_tokens=120, temperature=2.0).split()

pos = 0
for size in [random.randint(3, 9) for _ in range(20)]:
    chunk = words[pos:pos+size]
    if chunk:
        print(' '.join(chunk))
    pos += size
    if pos >= len(words):
        break
```

## eerie
order-3: looks back 3 words, so stoker's sentence structure mostly survives. it sounds like a journal entry but the story is gone, all the dread, none of the events.

```python
tok_eerie = SimpleWordTokenizer()
gen_eerie = TextGenerator(n=4, tokenizer=tok_eerie, source=sentences)

for _ in range(4):
    print(textwrap.fill(gen_eerie.generate_one(max_tokens=50, temperature=1.5), width=80))
    print()
```
