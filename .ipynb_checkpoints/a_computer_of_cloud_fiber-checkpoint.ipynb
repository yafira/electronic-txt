# Assignment #1: Poetry Generator

My generator produces poems about soft computers, imaginary machines made of fabric, powered gently, and designed without urgency. The vocabulary connects to my thesis work on gentler relationships between people and devices.

Each stanza follows the same seven-line cascade: material, assembly, power source, interface, location, inhabitant, promise. The indentation is intentional, each line steps further in, like something being carefully layered.

I based this on the simple random-vocabulary generators from the class notebook but made each stanza read as a complete object rather than loose lines. The vocabulary comes from my own design practice: conductive thread, industrial felt, "a screen that refuses urgency," "it will never ask for your attention twice."

Does it surprise me? Yes, mostly in the combinations. "A computer of tulle / taped with care (temporary) / powered by a warm battery" feels like something that could actually exist in a studio.

```python
import random
```

```python
# themed around soft interfaces + gentle tech, inspired by my thesis
materials = [
    "felt", "knit", "tulle", "conductive thread", "soft silicone", "foam",
    "velvet", "gel", "mesh", "industrial felt", "cloud fiber"
]
```

```python
structures = [
    "stitched with running thread",
    "sealed with a slow zipper",
    "laced with ribbon logic",
    "held by snaps and patience",
    "taped with care (temporary)",
    "woven in loops that remember"
]
```

```python
power_sources = [
    "moonlight", "usb power", "a warm battery", "static", "a dim wall outlet",
    "no power at all", "a shared charge"
]
```

```python
interfaces = [
    "a keyboard that blushes when touched",
    "a screen that refuses urgency",
    "buttons that only work if you're gentle",
    "a dial that drifts like weather",
    "a touchpad made of cloth and friction",
    "a cursor that waits for your breath"
]
```

```python
inhabitants = [
    "people who log off early",
    "hands learning softness",
    "a shy machine spirit",
    "tired students and their prototypes",
    "a small archive of care",
    "memories that don't want to be optimized"
]
```

```python
locations = [
    "under a desk lamp at midnight",
    "in a room where time is slower",
    "inside a tote bag",
    "by the window with quiet air",
    "in the studio after critique",
    "near a plant that survives anyway"
]
```

```python
promises = [
    "it will not rush you",
    "it will misbehave politely",
    "it will keep your secrets imperfectly",
    "it will take breaks with you",
    "it will glow only when needed",
    "it will never ask for your attention twice"
]
```

```python
stanza_count = 7

for i in range(stanza_count):
    print()
    print("A computer of " + random.choice(materials))
    print("     " + random.choice(structures))
    print("          powered by " + random.choice(power_sources))
    print("                with " + random.choice(interfaces))
    print("                     resting " + random.choice(locations))
    print("                          inhabited by " + random.choice(inhabitants))
    print("                               and " + random.choice(promises) + ".")
```
