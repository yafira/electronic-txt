# Assignment #2 Genre Study: The Horoscope

## Genre Analysis

I collected ~15 horoscopes from Astrology.com, Co-Star, and The Cut. Here are the **structural characteristics** common to all of them:

1. **Celestial cause → life effect**: A planet doing something *causes* a change in your personal life ("With Venus entering Taurus, your finances stabilize")
2. **Vague but specific-sounding advice**: Instructions that feel pointed but apply to everyone ("Now is the time to reflect on what you truly want")
3. **Emotional validation**: Acknowledges a feeling the reader *might* be having ("You may have felt drained lately")
4. **A dramatic warning**: Something to avoid, often tied to a planet in retrograde
5. **Closing affirmation**: Ends on an empowering or mystical note
6. **Zodiac-specific flavor**: Each sign gets associated personality traits woven in

The tone is: **authoritative, vague, cosmic, oddly personal, slightly ominous.**

```python
def generate(symbol, grammar):
    # if the symbol isn't in the grammar, return it as plain text
    if symbol not in grammar:
        return symbol

    # random rule for symbol
    text = random.choice(grammar[symbol])

    # replaces every {placeholder} by expanding it the same way
    for match in re.findall(r"\{(\w+)\}", text):
        text = text.replace("{" + match + "}", generate(match, grammar), 1)

    return text


def horoscope(sign=None):
    text = generate("origin", grammar)
    if sign:
        return f"✦ {sign.upper()} ✦\n{text}"
    return text


print("Gen ready!")
```

## Generate a single horoscope

```python
print(horoscope("Capricorn"))
```

## Generate horoscopes for all 12 signs

```python
signs = [
    "Aries", "Taurus", "Gemini", "Cancer", "Leo", "Virgo",
    "Libra", "Scorpio", "Sagittarius", "Capricorn", "Aquarius", "Pisces",
]

for sign in signs:
    print(horoscope(sign))
    print()
```
