```python
# can use for phonetic analysis in a text
from collections import Counter
text = "april is the cruelest month breeding lilacs out of the dead"
count = Counter()
words = text.split()
for item in words:
    pronounciations = pr.phones_for_word(item)
    if len(pronounciations) > 0:
        # print(item, pronounciations)
        phonemes = pronounciations[0].split()
        print(item, pronounciations, phonemes)
        # print(phonemes)
        count.update(phonemes)
    else:
        print("waring: can't pronounce", item)
count.most_common(5)
```
